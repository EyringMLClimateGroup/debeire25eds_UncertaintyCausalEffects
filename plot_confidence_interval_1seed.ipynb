```python
#Import useful packages
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import scienceplots
plt.style.use(['science','grid','no-latex'])
```

```python
def plot_results(df_res_up,df_res_low,df_res_est,causal_met,arg_seed,sample="large"):
    plt.clf()
    if sample == "large":
        df_res_up=df_res_up.iloc[4:,:]
        df_res_low=df_res_low.iloc[4:,:]
        df_res_est=df_res_est.iloc[4:,:]
    else:
        df_res_up=df_res_up.iloc[1:8,:]
        df_res_low=df_res_low.iloc[1:8,:]
        df_res_est=df_res_est.iloc[1:8,:]
    
    df_res_up.columns =["COV-Stand.","COV-Hall","RES-Stand.","RES-Hall","Asymp."]
    df_res_low.columns =["COV-Stand.","COV-Hall","RES-Stand.","RES-Hall","Asymp."]
    
    cmap = sns.color_palette("bright")
    sns.set_palette(cmap)
    sns.lineplot(data=df_res_up,lw=1.25)
    sns.lineplot(data=df_res_low,legend=False,lw=1.25)

    cmap=["black","grey"]
    sns.set_palette(cmap)
    
    axes = sns.lineplot(data=df_res_est,lw=1.25)
    axes.autoscale(tight=True)
    h,l = axes.get_legend_handles_labels()
    l1 = axes.legend(h[:5],l[:5], bbox_to_anchor=(1.00, 1),loc="upper left",title="CI method",fontsize=10, handlelength=1)
    l2 = axes.legend(h[5:],l[5:], bbox_to_anchor=(1.00, 0), loc="lower left",title="Sensitivity",fontsize=10, handlelength=1)
    axes.add_artist(l1)
    
    if "VAR" in causal_met:
        axes.text(-0.18, 1.1, "A", fontsize= 20, fontweight='bold', ha='left', va='top', transform=axes.transAxes)
    else:
        axes.text(-0.18, 1.1, "B", fontsize=20, fontweight='bold', ha='left', va='top', transform=axes.transAxes)
    # set the title
    axes.text(1.2,-.12,"Seed #%d"%arg_seed,fontsize= 10, fontweight='bold', ha='left', va='top', transform=axes.transAxes)
    plt.xlabel("Time sample size")
    plt.ylabel("Sensitivity")
    plt.title('%s - Sensitivity estimate and 90%s CI '%(causal_met,"%"),loc='left')
    plt.subplots_adjust(right=0.8)
    # show the plot and save it
    #ax1.autoscale(tight=True)
    save_filename="./plots/ci/ci_seed%d_%s%s.pdf"%(arg_seed,causal_met.split(" ")[-1],sample)
    plt.savefig(save_filename,dpi=600, bbox_inches="tight")
    print("Saved plot in %s"%save_filename)
    plt.show()
```
