```python
def plot_MAE(list_df_results,legend_label,title,y_label):
    #Get current axis
    n_res = len(list_df_results)
    plt.clf()
    fig, ax = plt.subplots()
    ax.set_prop_cycle(color=['#d62728', '#2ca02c'])
    for i,df in enumerate(list_df_results):
        ax.plot(df.index,df.MAE,linewidth=1.5)
    plt.legend(legend_label,title='Methods')
    # set the title
    plt.title('%s'%(title))
    plt.ylabel(y_label)
    plt.xlabel("Time sample size T")
    plt.savefig('./plots/ci/mae_comparison.pdf',dpi=600)
    print("Saved plot in %s"%'./plots/ci/mae_comparison.pdf')
    # show the plot
    plt.show()
    plt.close()
```

```python
plot_MAE([df_res_est_var,df_res_est_pcmci],["VAR model","PCMCI-CCM"],"Sensitivity mean absolute error (MAE)","MAE")
```

```python
def plot_ci(df,df_err,method_name):
    plt.clf()
    custom_cycler = cycler(color=sns.color_palette("bright"))
    plt.rc('axes', prop_cycle=custom_cycler)
    ax=df.plot(kind='bar',yerr=df_err.iloc[1:,:],width=0.75,title= "%s average CI size (90%s level)"%(method_name,"%"),ylabel="CI size",xlabel="Time sample size")
    if "VAR" in method_name:
        ax.text(-0.18, 1.1, "A", fontsize= 20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    else:
        ax.text(-0.18, 1.1, "B", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    plt.ylim([0,0.55])
    print("Saved plot in ./plots/ci/ci_comparison_%s.pdf"%method_name.split(" ")[0])
    plt.savefig('./plots/ci/ci_comparison_%s.pdf'%method_name.split(" ")[0],dpi=600)
    #ax.autoscale(tight=True)
    plt.show()
    plt.close()
```

```python
plot_ci(df_res_spread_var,df_res_spread_std_var,"VAR model")
```

```python
plot_ci(df_res_spread_pcmci,df_res_spread_std_pcmci,"PCMCI-CCM")
```

```python
def plot_ci_eval(df,method_name):
    plt.clf()
    custom_cycler = cycler(color=sns.color_palette("bright"))
    plt.rc('axes', prop_cycle=custom_cycler)
    ax = df.round(decimals=2).plot(kind='bar',legend=False,width=0.75,title= "%s - Empirical sig. level for 90%s CI"%(method_name,"%"),ylabel="Empirical sig. level",xlabel="Time sample size")
    plt.axhline(y=0.9, color='grey', linestyle='--',linewidth=1.5)
    if "VAR" in method_name:
        ax.text(-0.22, 1.1, "A", fontsize= 20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    else:
        ax.text(-0.22, 1.1, "B", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    plt.ylim([0.5, 1.])
    plt.savefig('./plots/ci/ci_empirical_eval_%s.pdf'%method_name.split(" ")[0],dpi=600)
    print("Saved plot in ./plots/ci/ci_empirical_eval_%s.pdf"%method_name.split(" ")[0])

    #ax.autoscale(tight=True)
    plt.show()
    plt.close()
```

```python
plot_ci_eval(df_res_ci_var,"VAR model")
```
