```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','grid','no-latex'])
```

```python
def plot_results(df_results,varying_paramater,title,y_label):
    #Get current axis
    sns.set_palette("tab10")
    ax = sns.lineplot(data=df_results, x="parameter", y="error", hue="method",linewidth=1.5,errorbar=('ci', 90))
    plt.legend(title='Methods')
    if varying_paramater == "Time sample size":
        ax.text(-0.15, 1.1, "A", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    if varying_paramater == "Number of modes":
        ax.text(-0.15, 1.1, "B", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    if varying_paramater == "Number of cross-links":
        ax.text(-0.15, 1.1, "C", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    if varying_paramater == "Auto-correlation strength":
        ax.text(-0.15, 1.1, "D", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    if varying_paramater == "Spatial resolution":
        ax.text(-0.15, 1.1, "E", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    if varying_paramater == "Covariant noise strength":
        ax.text(-0.15, 1.1, "F", fontsize=20, fontweight='bold', ha='left', va='top', transform=ax.transAxes)
    
    plt.title('%s %s'%(title,y_label))
    plt.ylabel(y_label)
    plt.xlabel(varying_paramater)
    plt.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))
    plt.savefig('./plots/benchmark/sens_err/%s_%s_%s.pdf'%(title,y_label,varying_parameter),dpi=600)
    plt.show()
    plt.close()
```

```python
def rename_method(df_):
    df_.loc[df_.method =="PCA_+PCMCI_+LinMed","method"]="PCA-PCMCI-CCM"
    df_.loc[df_.method =="Varimax+PCMCI_+LinMed","method"]="Varimax-PCMCI-CCM"
    df_.loc[df_.method =="PCA_+VARmodel","method"]="PCA-VAR"
    df_.loc[df_.method =="Varimax+VARmodel","method"]="Varimax-VAR"
    df_["method"]= pd.Categorical(df_.method,ordered=True)
    df_.sort_values("method")
    return df_
```

```python
varying_parameter= "T"
dfT_mrae=rename_method(pd.read_csv("./results/%s_LTE_MRAE.csv"%varying_parameter,index_col=0))
dfT_links=rename_method(pd.read_csv("./results/%s_Links_MSE.csv"%varying_parameter,index_col=0))
dfT_sens_err=rename_method(pd.read_csv("./results/%s_Sensitivity_Error.csv"%varying_parameter,index_col=0))
varying_parameter= "Time sample size"
plot_results(dfT_mrae,varying_parameter, "Long-run Effect","MRAE")
plot_results(dfT_links,varying_parameter,"Linear Coefficient","MSE")
plot_results(dfT_sens_err,varying_parameter,"Sensitivity","Absolute Error")
```
